```python
import pandas as pd
import folium
```

```python
df = pd.read_csv("densite-de-medecins.csv",sep=';')
df.dropna(inplace=True)
print(df)
```

```python
cols = ['Département','Spécialité','Densité']
df = df[cols]
df
```

```python
test = df[df['Spécialité'] == 'Généralistes'][['Département','Densité']].sort_values('Département')[:-25]
test.reset_index(inplace=True)
test = test[['Département','Densité']]
```

```python
test['Num_dep'] = test['Département'].apply(lambda x: x.split(' - ')[0])
test['Département'] = test['Département'].apply(lambda x: x.split(' - ')[1])
test
```

```python
import json
import os
state_geo = os.path.join('./france-geojson-master/departements.geojson')
#state_unemployment = os.path.join('data', 'US_Unemployment_Oct2012.csv')
#state_data = pd.read_csv(state_unemployment)

m = folium.Map(location=[48.856614, 2.352222], zoom_start=5)


m.choropleth(
    geo_data=state_geo,
    name='Densité de Médecin Généraliste par Département',
    data=test,
    columns=['Num_dep','Densité'],
    key_on='properties.code',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2
)


folium.LayerControl().add_to(m)

m
```

```python
data = pd.read_csv("rpps-medecins-tab7_51427175154982.csv",sep=',', skiprows=[0,1,2,3,5],encoding='latin-1')
data = data.rename(columns={"SPECIALITE" : "Region"})
print(data)
```

```python
newdata = pd.DataFrame()
newdata['Departement'] = data['Region'][21:]
newdata['Densite'] = data['Ensemble des spécialités d\'exercice'].apply(lambda x : x/(data[data['Region'] == 'FRANCE ENTIERE']['Ensemble des spécialités d\'exercice']))[21:]
newdata['Num_dep'] = newdata['Departement'].apply(lambda x: x.split(' - ')[0])
newdata['Departement'] = newdata['Departement'].apply(lambda x: x.split(' - ')[1])
newdata
```
